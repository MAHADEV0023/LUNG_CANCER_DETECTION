# SPLITING DATA

```python
X = cancer_data.drop('LUNG_CANCER',axis=1)
X
```

```python
Y = cancer_data['LUNG_CANCER']
Y
```

```python
from imblearn.over_sampling import RandomOverSampler

over_samp =  RandomOverSampler(random_state=0)
X_train_res, y_train_res = over_samp.fit_resample(X, Y)
X_train_res.shape, y_train_res.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_res, y_train_res, test_size = 0.2, random_state = 42)
```

```python
len(X_train), len(X_test)
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(10,6))
plt.title("No. of samples after balancing", fontsize=20, y=1.02)
sb.countplot(x = y_train_res)
plt.show()
```

```python
# Scaling the data 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# LinearRegression

```python

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
LinearRegressionScore = lr.score(X_test,y_test)
print("Accuracy of Linear Regression model:",LinearRegressionScore*100)
```

# Random Forest Classifier

```python

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100)
rfc.fit(X_train,y_train)
```

```python
RandomForestClassifierScore = rfc.score(X_test, y_test)
print("Accuracy of Random Forest Classifier model:",RandomForestClassifierScore*100)
```

```python
# Confusion Matrix of Random Forest Classifier
from sklearn.metrics import confusion_matrix, classification_report

y_pred_rfc = rfc.predict(X_test)
cf_matrix = confusion_matrix(y_test, y_pred_rfc)
sb.heatmap(cf_matrix, annot=True, cmap='viridis_r')
plt.title("Confusion Matrix for Random Forest Classifier", fontsize=20, fontname="Helvetica", y=1.03);
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred_rfc))
```

# K Neighbors Classifier

```python

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
KNeighborsClassifierScore = knn.score(X_test, y_test)
print("Accuracy of K Neighbors Classifier model:",KNeighborsClassifierScore*100)
```

```python
# Confustion Matrix of K Neighbors Classifier

y_pred_knn = knn.predict(X_test)
cf_matrix = confusion_matrix(y_test, y_pred_knn)
sb.heatmap(cf_matrix, annot=True, cmap='viridis_r')
plt.title("Confusion Matrix for K Neighbors Classifier", fontsize=20, fontname="Helvetica", y=1.03);
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
print(metrics.classification_report(y_test, y_pred_knn))
```
